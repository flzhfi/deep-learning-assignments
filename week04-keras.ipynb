```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier, KerasRegressor
import os
import numpy as np
```

```python
# 실험 결과 재현을 위한 seed 고정
SEED = 4
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
tf.random.set_seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
initializer = tf.keras.initializers.GlorotUniform(seed=SEED)
```

```python
# 데이터세트 준비
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
# 데이터 전처리
train_images = train_images.reshape((60000, 784)) # 2D 이미지 -> 벡터로 평탄화
train_images = train_images.astype('float32') / 255 # 정규화

test_images = test_images.reshape((10000, 784))
test_images = test_images.astype('float32') / 255

train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier, KerasRegressor
import os
import numpy as np

# 신경망 모델 구축
def build_model():
    network = tf.keras.models.Sequential()

    # 은닉층 생성
    # 여기에 코드를 작성해주세요
    network.add(tf.keras.layers.Dense(512, activation='relu', input_shape=(28*28,),
                                      kernel_initializer=initializer))
    network.add(tf.keras.layers.Dense(256, activation='relu',
                                      kernel_initializer=initializer))

    network.add(tf.keras.layers.Dense(10, activation='sigmoid',
                                      kernel_initializer=initializer))

    # 옵티마이저 설정 필요
    network.compile(optimizer='Adam', # Adam 사용
                   loss='mse',
                   metrics=['accuracy'])
    return network
```

```python
# model 생성
model = build_model()
```

```python
result = model.fit(train_images, train_labels, epochs=5, batch_size=64)
# epochs = 5
# batch_size = 64: 64개씩 묶어서
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```python
print('테스트 정확도 : ', test_acc)
```

### 가장 좋은 성능의 옵티마이저 작성
- 가장 좋은 성능: Adam

### 옵티마이저 서술
- RMSprop : 정확도가 Adam 다음으로 높게 나옴. 학습률이 0으로 떨어지지 않음.

- Adagrad : 정확도가 떨어짐. 학습이 진행될수록 학습률을 계속 줄이는 특성이 있음. 초반에는 정확도가 오르다가 중간에 정체되어 하락함.

- Adadelta : 학습 속도가 느리고, 성능이 좋지 못함. 학습률 파라미터가 아예 없음.

- Adam : 가장 빠르고 정확함. 현재 데이터셋에 적합하여 정확도가 제일 높게 나옴.

## 과제 2 MLP 모델의 정확도(Accuracy)를 71이상 달성하시오.  
- 해당 diabetes.csv 데이터는 Pima 인디언의 의료 기록과 각 환자가 5년 이내에 당뇨병 발병 여부가 저장되어 있다.
- 현재 모델은 1 : 당뇨병 양성, 0 : 당뇨병 음성을 분류하는 MLP 모델이다.
- 해당 모델의 하이퍼 파라미터를 수정하여 정확도 71 이상을 달성하여라.
- 하이퍼 파라미터 : Activation Function, optimizer, epochs, batch size 등

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad, Adadelta, RMSprop, Adam
import pandas as pd
import numpy as np
import os
```

```python
# 데이터 load
dataset = pd.read_csv(os.path.join(data_path, 'diabetes.csv'))

X = dataset.iloc[:, 0:8].values
y = dataset.iloc[:, 8].values
```

```python
# 수정 x
def configure_tf_determinism():
    # TensorFlow의 병렬성 감소
    tf.config.experimental.enable_op_determinism()
```

```python
# 실험 결과 재현을 위한 seed 고정
SEED = 4
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
tf.random.set_seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
initializer = tf.keras.initializers.GlorotUniform(seed=SEED)
```

```python
# 모델 생성
def build_model():
    model = Sequential()
    # 수정 가능
    # 활성화함수: relu 사용
    model.add(tf.keras.layers.Dense(12, input_dim=8, activation='relu', kernel_initializer=initializer))
    model.add(tf.keras.layers.Dense(8, activation='relu', kernel_initializer=initializer))

    # 수정 x
    model.add(tf.keras.layers.Dense(1, activation='sigmoid', kernel_initializer=initializer))

    # 옵티마이저: Adam으로 수정
    model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])

    return model
```

```python
# 수정 x
configure_tf_determinism()
```

```python
model = build_model()
```

```python
model.fit(X, y, epochs=10, batch_size=1)
```

```python
_, accuracy = model.evaluate(X, y)
```

```python
print('정확도: %.2f' % (accuracy*100))
```
